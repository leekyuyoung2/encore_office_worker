```python
import pandas as pd
df = pd.read_csv('seoul.csv')
# 상권코드명
```

```python
office_df = pd.read_csv('상권명_주소매핑.csv')  # 상권명
```

```python
seoul_df = pd.merge(df,office_df,how='left',left_on='상권코드명',right_on='상권명')
seoul_df.head(2)
```

```python
# 결측치 제거
not_null_seroul_df = seoul_df.dropna(subset=['address_name'])
(not_null_seroul_df.isna().sum())[not_null_seroul_df.isna().sum()>0]
```

```python
not_null_seroul_df['행정구역명'] = not_null_seroul_df['address_name'].apply(lambda x : x.split()[1])
```

```python
not_null_seroul_df.columns = not_null_seroul_df.columns.str.replace('당월매출금액','월매출금액')
not_null_seroul_df.columns = not_null_seroul_df.columns.str.replace('당월매출건수','월매출건수')
```

```python
not_null_seroul_df.to_csv('clean_seoul.csv')
```
