```python
import requests
url = 'https://news.naver.com/'
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html,'html.parser')
h3_tags = soup.find_all('h3')
# print(h3_tags[0].text)

h3_tags # 리스트  -- for 순환문을 이용하면 요소를 순환하면서 각 요소를 추출할수 있다

for tag in h3_tags:    
    print(tag.text.strip())
```

```python
import requests
url = 'https://news.naver.com/'
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')

# class 속성으로 찾기 (주의: class_로 쓴다)
articles = soup.find_all('span', class_='Nitem_link_menu')
print(len(articles))  # 출력: 2
```

```python
for article in articles:
    print(article.text)
```

```python
for i in range(6):
    print('*'*50)
    pagenum = 100 + i
    url = f'https://news.naver.com/section/{pagenum}'
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    # 뉴스목록을 가져와 <ul>  <li>  <li>태그들을 순환해서 정보 출력
    news = soup.find('ul', class_='sa_list')
    li_lists = news.find_all('li')

    for li in li_lists:
        title = li.find('strong', class_='sa_text_strong').text
        article = li.find('div', class_='sa_text_lede').text
        print(f'title:{title}  article:{article}')
```

```python
import pandas as pd
# CSS 선택자 select로 비교
url = f'https://news.naver.com/section/100'
headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url,headers=headers)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
headlines = soup.select('#newsct > div.section_component.as_section_headline._PERSIST_CONTENT> div.section_article.as_headline._TEMPLATE > ul > li')

data = []
for li in headlines:    
    title = li.find('strong', class_='sa_text_strong').text
    article = li.find('div', class_='sa_text_lede').text
    press = li.select_one('.sa_text_press').text    
    link = li.select('a[data-imp-url]')
    # print(f'title:{title}  article:{article}')
    data.append([title,article,press,link])
df = pd.DataFrame(data,columns=['제목', '내용','언론사','링크'])
df.to_csv('정치해드라인.csv', encoding='utf-8-sig', index=False)
df
```
