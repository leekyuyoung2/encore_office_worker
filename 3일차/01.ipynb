```python
# CSV 다루기  - pandas
# json 데이터 - 중첩된 dictionary를 문자열로 표현
# 문자열 -> 파이썬 딕셔너리/ 리스트   파싱(변환)
# 파이썬 딕셔너리 /리스트 -> 문자열   덤프(반대)
import json
json_string =  '''
{
    "고객":{
        "이름":"김철수",
        "나이":30
    },
    "다른고객":{
        "이름":"홍길동",
        "나이":35
    }
}
'''
data = json.loads(json_string)  # 문자열 -> 객체
data['고객']
```

```python
# 딕셔너리 -> 문자열
output_json = json.dumps(data['고객'],ensure_ascii=False,indent=4)
print(output_json)
```
